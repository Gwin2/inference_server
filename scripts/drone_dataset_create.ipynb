```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#classes = {'noise': 0, 'drone': 1, 'control': 2, 'wifi': 3}
classes = ['0','1','2','3']
total_ctrs = {0: 0, 1: 0, 2: 0, 3: 0}
```

```python
path = 'S:/drone_dataset_v4.0/'
```

```python
dirs = os.listdir(path)
```

```python
dirs
```

```python
pd_columns = ['fragment_name', 'label']
df = pd.DataFrame(columns=pd_columns)
```

```python
for dir in dirs:
    files = os.listdir(path + dir)
    for file in files:
        s = dir[0]
        if s in classes:
            s = int(s)
            total_ctrs[s] += 1
            print(file, s)
            row = pd.DataFrame({pd_columns[0]: [dir + '/' + file], pd_columns[1]: s})
            df = pd.concat([df, row], ignore_index=True)
```

```python
df.to_csv(path + 'dataset.csv', index=False)
```
