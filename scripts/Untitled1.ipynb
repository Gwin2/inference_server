```python
import numpy as np
import pandas as pd
import os
import random
import time
import matplotlib.pyplot as plt

root = ''
path = root + 'dataset2/'
files = os.listdir(path)
with open(path + files[5], 'rb') as rfl:
  sig = np.frombuffer(rfl.read(), dtype=np.complex64)
  plt.figure(figsize=(14,8))
  plt.plot(sig.real)
try:
  os.mkdir(root+'calib')
except:
  pass
with open(root + 'calib/dataset.txt', 'w') as wfl:
    for file in files:
        if not file.endswith('.npy'):
          continue
        with open(path + file, 'rb') as rfl:
            sig = np.frombuffer(rfl.read(), dtype=np.complex64)
        print(sig.size)
        print('calib/' + file)
        np.save(root + 'calib/' + file, sig)
        wfl.write('calib/' + file + '\n')
```
